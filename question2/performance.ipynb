2. Gender vs Test Performance (Box plot)

3. Parental Education vs Test Scores (Bar chart)

```python
# Parental Education vs Test Scores
plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='math score', data=df, palette='muted')
plt.xticks(rotation=45)
plt.title('Parental Level of Education vs Math Scores')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='reading score', data=df, palette='muted')
plt.xticks(rotation=45)
plt.title('Parental Level of Education vs Reading Scores')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='writing score', data=df, palette='muted')
plt.xticks(rotation=45)
plt.title('Parental Level of Education vs Writing Scores')
plt.show()
```

4. Lunch Type vs Test Scores (Scatter plot)
